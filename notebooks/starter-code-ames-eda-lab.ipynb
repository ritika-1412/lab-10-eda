## Part 3 — Missing Values

Before analyzing relationships, check for missing values.

Then fill in the missing values with an appropriate value.

**DO NOT HARDCOSE THE VALUE TO FILL THE MISSING VALUE**

Use `.fillna()`

```python
# Count missing values per column
housing.isnull().sum()
```

```python
# Show missing value percentage per column
missing_pct = housing.isnull().mean() * 100
missing_pct[missing_pct > 0]
```

```python
numeric_cols = housing.select_dtypes(include='number').columns
categorical_cols = housing.select_dtypes(exclude='number').columns

print('Numeric columns:', list(numeric_cols))
print('Categorical columns:', list(categorical_cols))
```

```python
# Fill numeric columns with the median of each column
for col in numeric_cols:
    housing[col] = housing[col].fillna(housing[col].median())

# Fill categorical columns with "Missing"
for col in categorical_cols:
    housing[col] = housing[col].fillna("Missing")

# Verify no missing values remain
housing.isnull().sum()
```

### Reflection

- Why might the median be safer than the mean for numeric imputation?

## Part 4 — Summary Statistics for Numeric Variables

Explore individual numeric variables.

Manually caluclate the mean and median and then compare your calculation with Pandas.

```python
# Manual mean: sum of all values divided by count
manual_mean = housing['Sale_Price'].sum() / len(housing['Sale_Price'])
print(f"Manual mean Sale_Price:  {manual_mean:,.2f}")
```

```python
# Manual median: sort values and pick the middle element
sorted_prices = sorted(housing['Sale_Price'])
n = len(sorted_prices)
if n % 2 == 1:
    manual_median = sorted_prices[n // 2]
else:
    manual_median = (sorted_prices[n // 2 - 1] + sorted_prices[n // 2]) / 2

print(f"Manual median Sale_Price: {manual_median:,.2f}")
```

```python
# Compare manual calculations with Pandas
pandas_mean   = housing['Sale_Price'].mean()
pandas_median = housing['Sale_Price'].median()

print(f"Pandas mean Sale_Price:   {pandas_mean:,.2f}")
print(f"Pandas median Sale_Price: {pandas_median:,.2f}")
print(f"\nMean matches:   {abs(manual_mean   - pandas_mean)   < 0.01}")
print(f"Median matches: {abs(manual_median - pandas_median) < 0.01}")
```

```python
# Summary statistics for all numeric columns
housing[numeric_cols].describe()
```

```python
# Range of Sale_Price
price_range = housing['Sale_Price'].max() - housing['Sale_Price'].min()
print(f"Sale_Price range: {price_range:,}")
print(f"Sale_Price min:   {housing['Sale_Price'].min():,}")
print(f"Sale_Price max:   {housing['Sale_Price'].max():,}")
```

```python
# Manual calculation: population variance and standard deviation

mean_price = housing['Sale_Price'].mean()
manual_pop_var = ((housing['Sale_Price'] - mean_price) ** 2).sum() / len(housing['Sale_Price'])
manual_pop_std = manual_pop_var ** 0.5

print(f"Manual population variance: {manual_pop_var:,.2f}")
print(f"Manual population std dev:  {manual_pop_std:,.2f}")
```

```python
# Compare with Pandas (sample std, ddof=1 by default)
pandas_sample_std = housing['Sale_Price'].std()
pandas_sample_var = housing['Sale_Price'].var()

print(f"Pandas sample std dev:      {pandas_sample_std:,.2f}")
print(f"Pandas sample variance:     {pandas_sample_var:,.2f}")
print(f"\n(Population std is slightly smaller than sample std because denominator is n vs n-1)")
```

## Part 6 — Grouped Summary Statistics

Use `.groupby()` to compare sale prices across categories.


```python
# Mean and median Sale_Price by Neighborhood, sorted by mean
neighborhood_stats = (
    housing.groupby('Neighborhood')['Sale_Price']
    .agg(mean='mean', median='median', count='count')
    .sort_values('mean', ascending=False)
)
print(neighborhood_stats.to_string())
```

```python
# Mean and median Sale_Price by House_Style
house_style_stats = (
    housing.groupby('House_Style')['Sale_Price']
    .agg(mean='mean', median='median', count='count')
    .sort_values('mean', ascending=False)
)
print(house_style_stats.to_string())
```

```python
# Mean and median Sale_Price by Central_Air
central_air_stats = (
    housing.groupby('Central_Air')['Sale_Price']
    .agg(mean='mean', median='median', count='count')
)
print(central_air_stats)
```

```python
# Mean and median Sale_Price by Garage_Type
garage_stats = (
    housing.groupby('Garage_Type')['Sale_Price']
    .agg(mean='mean', median='median', count='count')
    .sort_values('mean', ascending=False)
)
print(garage_stats.to_string())
```

## Part 7 — Covariance and Correlation

Use covariance and correlation to investigate relationships between numeric variables.


```python
# Covariance matrix for numeric variables
cov_matrix = housing[numeric_cols].cov()
cov_matrix
```

```python
# Correlation matrix (Pearson) — easier to interpret than covariance
corr_matrix = housing[numeric_cols].corr()
corr_matrix
```

```python
# Correlations with Sale_Price, sorted by absolute strength
sale_price_corr = corr_matrix['Sale_Price'].drop('Sale_Price').sort_values(key=abs, ascending=False)
print("Correlation with Sale_Price:")
print(sale_price_corr)
```

```python
# Manual covariance between Gr_Liv_Area and Sale_Price for verification
n = len(housing)
mean_area  = housing['Gr_Liv_Area'].mean()
mean_price = housing['Sale_Price'].mean()
manual_cov = ((housing['Gr_Liv_Area'] - mean_area) * (housing['Sale_Price'] - mean_price)).sum() / (n - 1)

pandas_cov = housing['Gr_Liv_Area'].cov(housing['Sale_Price'])

print(f"Manual covariance (Gr_Liv_Area, Sale_Price): {manual_cov:,.2f}")
print(f"Pandas covariance (Gr_Liv_Area, Sale_Price): {pandas_cov:,.2f}")
print(f"Match: {abs(manual_cov - pandas_cov) < 0.01}")
```

## Part 8 — Pandas Plots

Use Pandas plotting methods to support your analysis.


```python
import matplotlib.pyplot as plt

# Histogram of Sale_Price to see its distribution
housing['Sale_Price'].plot(
    kind='hist',
    bins=40,
    edgecolor='white',
    title='Distribution of Sale Price'
)
plt.xlabel('Sale Price ($)')
plt.ylabel('Count')
plt.tight_layout()
plt.show()
```

```python
# Scatter plot: Above-ground living area vs. Sale Price
housing.plot(
    kind='scatter',
    x='Gr_Liv_Area',
    y='Sale_Price',
    alpha=0.3,
    title='Living Area vs. Sale Price'
)
plt.xlabel('Above-Ground Living Area (sq ft)')
plt.ylabel('Sale Price ($)')
plt.tight_layout()
plt.show()
```

```python
# Bar chart: median Sale_Price by Central_Air
housing.groupby('Central_Air')['Sale_Price'].median().plot(
    kind='bar',
    title='Median Sale Price by Central Air',
    color=['steelblue', 'coral'],
    edgecolor='white'
)
plt.xlabel('Central Air (Y = Yes, N = No)')
plt.ylabel('Median Sale Price ($)')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()

# Box plot: Sale_Price distribution by House_Style
housing.boxplot(column='Sale_Price', by='House_Style', figsize=(10, 5), rot=30)
plt.suptitle('')
plt.title('Sale Price Distribution by House Style')
plt.xlabel('House Style')
plt.ylabel('Sale Price ($)')
plt.tight_layout()
plt.show()
```

## Final Written Analysis

### 1. What factors appear most related to sale price?

The strongest predictors of sale price are **above-ground living area (`Gr_Liv_Area`)** and **year built (`Year_Built`)**. Both have the highest positive correlations with `Sale_Price` among the numeric variables. Among categorical variables, **neighborhood** shows the largest spread in median sale prices — some neighborhoods (e.g., Stone_Brook, Northridge_Heights) have medians well above $300,000, while others fall below $100,000. **Central air conditioning** also shows a meaningful difference, with homes that have it selling for substantially more on average.

### 2. What evidence supports your answer?

- The correlation table shows `Gr_Liv_Area` has the highest positive correlation with `Sale_Price` (approx 0.71), followed by `Year_Built` (approx 0.56).
- The scatter plot of `Gr_Liv_Area` vs. `Sale_Price` shows a clear upward trend.
- The grouped summary for `Neighborhood` reveals median prices ranging from under $100k to over $300k depending on location.
- Homes with central air have a noticeably higher median sale price than those without.

### 3. What surprised you?

`Lot_Area` has a weaker correlation with sale price than expected. Intuitively, a bigger lot might mean a more expensive home, but the data suggests that living area matters far more than land area.

### 4. What would you investigate next?

- The **interaction between neighborhood and year built** could reveal whether newer homes command a premium only in certain locations.
- **Outlier investigation**: a handful of very expensive or very cheap homes may be distorting the mean and correlation calculations.

### 5. What limitations should we keep in mind?

- **Correlation ≠ causation.** A larger living area is associated with higher price, but we cannot say that adding square footage *causes* a price increase without controlling for other factors.
- The dataset covers Ames, Iowa from 2006–2010. Findings may not generalize to other cities, markets, or time periods.
- We examined only 10 of the 75 available columns; important predictors may have been excluded from this analysis.
