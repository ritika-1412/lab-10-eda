## Part 2 — Focus the Dataset

The full dataset has many columns. For this lab, we will focus on a smaller set of variables.


```python
columns_to_use = [
    'Sale_Price',
    'Gr_Liv_Area',
    'Overall_Qual',
    'Year_Built',
    'Lot_Area',
    'Lot_Frontage',
    'Neighborhood',
    'House_Style',
    'Garage_Type',
    'Central_Air'
]

housing = ames[columns_to_use].copy()
housing.head()
```

```python
housing.info()
```

## Part 3 — Missing Values

Before analyzing relationships, check for missing values.


```python
housing.isna().sum().sort_values(ascending=False)
```

```python
missing_pct = housing.isna().mean().sort_values(ascending=False) * 100
missing_pct
```

```python
numeric_cols = housing.select_dtypes(include='number').columns
categorical_cols = housing.select_dtypes(exclude='number').columns

print('Numeric columns:', list(numeric_cols))
print('Categorical columns:', list(categorical_cols))
```

```python
housing_clean = housing.copy()

for col in numeric_cols:
    housing_clean[col] = housing_clean[col].fillna(housing_clean[col].median())

for col in categorical_cols:
    housing_clean[col] = housing_clean[col].fillna('Missing')

housing_clean.isna().sum()
```

### Reflection

- Which columns had missing values?
- Why might the median be safer than the mean for numeric imputation?
- Why might we use `'Missing'` instead of dropping rows for categorical variables?


## Part 4 — Summary Statistics for Numeric Variables

Explore individual numeric variables.


```python
housing_clean[numeric_cols].describe()
```

```python
sale_prices = housing_clean['Sale_Price']

manual_mean = sale_prices.sum() / len(sale_prices)
manual_mean
```

```python
sale_prices.mean()
```

```python
sorted_prices = sale_prices.sort_values().reset_index(drop=True)
n = len(sorted_prices)

if n % 2 == 1:
    manual_median = sorted_prices[n // 2]
else:
    manual_median = (sorted_prices[(n // 2) - 1] + sorted_prices[n // 2]) / 2

manual_median
```

```python
sale_prices.median()
```

```python
# Manual calculation: population variance and standard deviation
# Note: Pandas uses sample standard deviation by default, so results may differ slightly.

mean_price = sale_prices.mean()
squared_diffs = (sale_prices - mean_price) ** 2
population_variance = squared_diffs.mean()
population_stdev = population_variance ** 0.5

population_variance, population_stdev
```

```python
sale_prices.var(), sale_prices.std()
```

### Reflection

- Is the mean sale price higher or lower than the median?
- What might that tell you about the distribution of sale prices?
- Is there a wide spread in sale prices?
